```python
import pandas as pd
import numpy as np

df_OWG = pd.read_excel('OneWord_Game.xlsx', sheet_name=None)
```

```python
df_all = pd.concat(df_OWG.values(), axis=1)
df_OWG_text = df_all.loc[13]
df_OWG_text
```

```python
df_sheets = []
for sheet_name, df in df_OWG_text.items():
  df_sheets.append(df)
  print(f'{sheet_name}, {type(df)}')
  pass
```

```python
text = ' '.join(map(str, df_sheets))
text
```

```python
# 조사 빼내기
import re
patterns = '(장영지|다양한|하거나|해야할|으로|에서|에는|위해|이를|이나|이다|있다|있게|되어|되고|되는|된다|까지|하지|하여|하고|하기|하게|하면|하며|하는|해야|한다|한지|로서|등 |할 |수 |서 |로 |때 |해 |의 |을 |를 |은 |과 |와 |에 |이 |가 |것|그 |\.|\,|\\n)'
text_regex = re.sub(pattern=patterns, repl=' ', string=text)
text_regex
```

```python
# !apt-get -y install fonts-nanum
# !fc-cache -fv
# !rm ~/.cache/matplotlib -rf
```

```python
from wordcloud import WordCloud
from collections import Counter
```

```python
text_split = text_regex.split()
text_split_count = Counter(text_split)
```

```python
top_30 = pd.Series(text_split_count).sort_values(ascending=False)[:30]
```

```python
wordCloud = WordCloud(background_color='white', font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf')
wordCloud.generate_from_frequencies(top_30)
```

```python
wordCloud.words_
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordCloud)
plt.axis('off')
plt.show()
```
