```python
import pandas as pd
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
# 데이터 로드
data = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = data.data
```

```python
documents[:1]
```

### TfidfVectorizer
max_df=, min_df=

```python
tfidfVectorizer = TfidfVectorizer(stop_words='english', max_df=0.95, min_df=2)
```

```python
tfidf = tfidfVectorizer.fit_transform(documents)
tfidf.toarray()[:1]
```

```python
import numpy as np
np.where(tfidf.toarray()[:1] > 0)
```

```python
latentDirichletAllocation3 = LatentDirichletAllocation(n_components=3, n_jobs=-1)
```

```python
latentDirichletAllocation3.fit(tfidf)
```

```python
vis = pyLDAvis.lda_model.prepare(latentDirichletAllocation3, tfidf, tfidfVectorizer)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis)
```
